### Baseline model 1

```python
import numpy as np
```

```python
class BaseLineModelPreviousHour:
    """Prediction = actual demand observed in the last hour
    """
    
    def fit(self, X_train: pd.DataFrame, y_train: pd.Series):
        """no training required for base line model

        Args:
            X_train (pd.DataFrame): _description_
            y_train (pd.Series): _description_
        """
        pass
    
    
    def predict(self, X_test: pd.DataFrame) -> np.ndarray:
        """Predict the demand for the next hour based on the actual demand observed in the last hour.

        Args:
            X_test (pd.DataFrame): _description_

        Returns:
            np.ndarray: _description_
        """
        
        return X_test[f"rides_previous_1_hour"]
```

### Create object of the base line class

```python
base_model = BaseLineModelPreviousHour()
predictions = base_model.predict(X_test)
predictions
```

```python
X_test["rides_previous_1_hour"]
```

### Error Metric: MAE

```python
from sklearn.metrics import mean_absolute_error
```

```python
test_mae = mean_absolute_error(y_test, predictions)
print(f"test MAE: {test_mae:.4f}")
```

### Baseline Model 2 : observe the demand from last week

```python
class BaselinModelPreviousWeek:
    """Prediction = actual demand observed at t - 7 days (observe the last week data to predict)
    """
    
    def fit(self, X_train: pd.DataFrame, y_train: pd.Series):
        pass
    
    
    def predict(self, X_test: pd.DataFrame) -> np.ndarray:
        return X_test[f"rides_previous_{7 * 24}_hour"]
```

```python
model_2 = BaselinModelPreviousWeek()
predictions = model_2.predict(X_test)
```

```python
test_mae = mean_absolute_error(y_test, predictions)
print(f"Baseline model 2---> test MAE: {test_mae:.4f}")
```

2nd baseline model is an improvement over the 1st baseline

### Baseline Model 3

```python
class BaselineModelLast4Weeks:
    """ 
    Prediction = actual deman observed at t - 7 days, t - 14 days, t - 21 days, t - 28 days
    """
    
    def fit(self, X_train: pd.DataFrame, y_train: pd.Series):
        pass
    
    
    def predict(self, X_test: pd.DataFrame) -> pd.Series:
        
        """take avg of the last 4 weeks demand to predict the next hour demand
        """
        return 0.25 * (
            X_test[f"rides_previous_{7 * 24}_hour"] + \
            X_test[f"rides_previous_{2 * 7 * 24}_hour"] + \
            X_test[f"rides_previous_{3 * 7 * 24}_hour"] + \
            X_test[f"rides_previous_{4 * 7 * 24}_hour"]
        )
    
    
```

```python
model_3 = BaselineModelLast4Weeks()
predictions = model_3.predict(X_test)
```

```python
test_mae = mean_absolute_error(y_test, predictions)
print(f"Baseline model 3---> test MAE: {test_mae:.4f}")
```

So far, baseline model 3 gave the best MAE score. So, now we can compare the ML models with this baseline score
